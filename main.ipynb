```python
# Version Control Skills Lab program
# Takes a name + a number, then prints a personalized result.

name = input("Enter your name: ").strip()

num_text = input("Enter a whole number (1-20): ").strip()
while not num_text.isdigit() or not (1 <= int(num_text) <= 20):
    num_text = input("Please enter a whole number from 1 to 20: ").strip()

n = int(num_text)

bar = "█" * n + "░" * (20 - n)
print(f"\nHello, {name}!")
print(f"Your power level: [{bar}] {n}/20")

print("\nYour custom chant:")
for i in range(n):
    print(f"{i+1}. {name} is locked in 💻")
```
